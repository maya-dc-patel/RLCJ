```python
# Evaluation metrics for the model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys

import statsmodels.api as sm

# X is total time steps
# fps indicator of how things it did a second

# explained variance shows how much of the variance in the data is explained by the model

# policy loss is the loss of the policy network 

# n updates is the number of updates to the model logging the metrics

# value loss is the loss of the value network

# entropy is the entropy of the policy distribution

# learning rate 

# total timesteps is the total number of timesteps

# time elapsed is the time elapsed

# total time is the total time

# std is the standard deviation
```

```python
# A2C policy loss vs total timesteps

df_A2C = pd.read_csv('A2C/progress.csv')
df_A2C.head()

# get the columns and put them in a list
columns = df_A2C.columns.tolist()
columns

lowess = sm.nonparametric.lowess
z = lowess(df_A2C['train/policy_loss'], df_A2C['time/total_timesteps'], frac=0.01,)
plt.plot(z[:,0], z[:,1], label='lowess fit', color='red')
#plt.scatter(df_A2C['time/total_timesteps'], df_A2C['train/policy_loss'], s= 1, color='blue', label='Policy Loss')
# smaller points in the scatter plot


# plot a line graph of the total timesteps vs policy loss
#plt.plot(df_A2C['time/total_timesteps'], df_A2C['train/policy_loss'], color='blue', label='Policy Loss')
plt.xlabel('Total Timesteps')
plt.ylabel('Policy Loss')
plt.title('A2C Policy Loss vs Total Timesteps')
plt.show()

```

```python
# A2C value loss vs total timesteps

# plot a line graph of the total timesteps vs value loss
plt.plot(df_A2C['time/total_timesteps'], df_A2C['train/value_loss'], color='blue', label='Value Loss')
plt.xlabel('Total Timesteps')
plt.ylabel('Value Loss')
plt.title('A2C Value Loss vs Total Timesteps')
plt.show()

```

```python
# A2C entropy loss vs total timesteps

# plot a line graph of the total timesteps vs entropy loss
plt.plot(df_A2C['time/total_timesteps'], df_A2C['train/entropy_loss'], color='blue', label='Entropy Loss')
plt.xlabel('Total Timesteps')
plt.ylabel('Entropy Loss')
plt.title('A2C Entropy Loss vs Total Timesteps')
plt.show()
```

```python
# A2C explained variance vs total timesteps

# plot a line graph of the total timesteps vs explained variance
plt.plot(df_A2C['time/total_timesteps'], df_A2C['train/explained_variance'], color='blue', label='Explained Variance')
plt.xlabel('Total Timesteps')
plt.ylabel('Explained Variance')
plt.title('A2C Explained Variance vs Total Timesteps')
plt.show()
```

```python
# A2c fps vs total timesteps

# plot a line graph of the total timesteps vs fps
plt.plot(df_A2C['time/total_timesteps'], df_A2C['time/fps'], color='blue', label='Frames Per Second')
plt.xlabel('Total Timesteps')
plt.ylabel('Frames Per Second (FPS)')
plt.title('A2C FPS vs Total Timesteps')
```
